## ‚ÄúLas 20 Peliculas y series mayor vistas en Disney Plus 2021‚Äù üé¨

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importaci√≥n de nuestro CSV ‚úÖ

```python
ds_disney_p = pd.read_csv('./disney_plus_titles.csv')
ds_disney_p.head(2)
```

```python
ds_disney_p_Es= pd.read_csv('./disney_plus_titles.csv', sep=',', header=0, names=['Tipo', 'Titulo', 'Director', 'Reparto', 'Pa√≠s',
       'Fehca en Disney+','Estreno', 'Transmici√≥n', 'Duraci√≥n (Minutos)','Clasifiaci√≥n','Descripci√≥n'],index_col=0)
ds_disney_p_Es_vi = ds_disney_p_Es.reset_index(drop=False)
ds_disney_p_Es_vi.head(2)
```

* Recopilaci√≥n de los 20 Titulos m√°s vistos ‚úÖ 

```python
df_disney_p_Es_20= ds_disney_p_Es_vi.head(20)
df_disney_p_Es_20.head(2)
```

### Ahora que ya tenemos nuestro DF listo,**resolveremos una serie de preguntas** que nos ayudar√° a nuestro **Analisis de Datos**
**1.** ¬øDe los titulos m√°s vistos en Disney+ 2021, qu√© tipo de **contendio** es? üßê

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_disney_p_Es_20, x='Tipo', y='Titulo');
plt.show()
```

2. Ademas de su **tipo** quiero saber ¬øDonde fue **producido** este contenido? üåé

```python
plt.figure(figsize=(8,6.2))
plt.title('Relaci√≥n entre los 20 Titulos m√°s vistos en Disney+ y su tipo ')
sns.swarmplot(data = df_disney_p_Es_20, x ='Tipo',y = 'Titulo', hue= 'Pa√≠s', dodge= False, marker = 'p', size=10);
plt.legend(bbox_to_anchor=(0.9, 0.9), loc='upper left', bbox_transform=plt.gcf().transFigure)
```

**3.** Excelente, pero quiero tener una **ayuda visual** que me permita entnder mejor **las categorias m√°s vistas en este DF**

* ### Representacion en una grafica de Pastel üç∞

```python
# Obtener el recuento de cada categor√≠a
counts = df_disney_p_Es_20['Tipo'].value_counts()

# Creamos una lista de etiquetas, valores & Colores
labels = counts.index
values = counts.values
colors = ['#3D8361', '#C7E8CA']
```

```python
# Creando un gr√°fico de pastel utilizando Matplotlib
plt.pie(values, labels=labels, autopct='%1.1f%%', colors= colors)

# Aspecto del gr√°fico de pastel

title_font = {'fontsize': 23, 'fontweight': 'bold', 'color': '#454545'}
plt.title("Distribuci√≥n de Categor√≠as", fontdict=title_font)

plt.axis('equal')
plt.show()
```

**4**. Ahora me interesa saber ¬ø**Qu√© relaci√≥n** existe entre el **a√±o de estreno** y **su duraci√≥n** de este contenido? 

```python
# Convertimos la columna 'Duraci√≥n (Minutos)' a tipo num√©rico
df_disney_p_Es_20['Duraci√≥n (Minutos)'] = pd.to_numeric(df_disney_p_Es_20['Duraci√≥n (Minutos)'])

# Ordenamos el DF de la columna 'Duraci√≥n (Minutos)' de menor a mayor
df_disney_p_Es_20_sorted = df_disney_p_Es_20.sort_values('Duraci√≥n (Minutos)', ascending=True)

plt.figure(figsize=(8, 6))
plt.title('Relaci√≥n entre A√±o de Estreno vs Duraci√≥n')
sns.scatterplot(data=df_disney_p_Es_20_sorted, x='Estreno', y='Duraci√≥n (Minutos)', hue='Titulo', palette='colorblind');
plt.xlabel('A√±o de Estreno')
plt.ylabel('Duraci√≥n en Minutos')
plt.legend(bbox_to_anchor=(0.9, 0.9), title='Titulo',loc='upper left', bbox_transform=plt.gcf().transFigure)
plt.show()
```

**5**. Por ultimo me interesa saber ¬øCual es la **clasifiacion es m√°s exitosa** en este DF? R = **Animation, Comedy, Family**

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df_disney_p_Es_20, x='Tipo', hue='Clasifiaci√≥n')

plt.title('Distribuci√≥n de Clasificaci√≥n por Tipo de Contenido')
plt.xlabel('Tipo de Contenido')
plt.ylabel('Cantidad')
# plt.legend(title='Clasificaci√≥n')
plt.legend(bbox_to_anchor=(0.9, 0.9), title='Clasificaci√≥n',loc='upper left', bbox_transform=plt.gcf().transFigure)
plt.show()
```
